```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Models and their respective RMAE values for supply and demand
models =     ["LSTM", "CNN", "HYBRID", "DT", "RDF"]
supply_rmse = [35.63, 40.16, 48.22, 22.52, 20.51]  # Example RMAE values for supply
demand_rmse = [30.03, 38.23, 23.89, 10.23, 10.89]  # Example RMAE values for demand

# Bar width and positions
bar_width = 0.35
x = np.arange(len(models))

# Create bar chart
plt.figure(figsize=(15, 6))
plt.bar(x - bar_width/2, supply_rmse, bar_width, label="Supply RMAE", color="#87CEFA")
plt.bar(x + bar_width/2, demand_rmse, bar_width, label="Demand RMAE", color="#FFA07A")


# Labels and title
plt.xlabel("Models")
plt.ylabel("RMAE")
plt.title("Supply vs Demand Model RMAE Test Results Comparison")
plt.xticks(x, models, fontsize=14)  # Set x-axis labels
plt.ylim(0, 50)  # Set y-axis to a logarithmic scale
plt.legend(["Supply", "Demand"], loc="upper left")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show values on top of bars
for i in range(len(models)):
    plt.text(x[i] - bar_width/2, supply_rmse[i] + 0.02, f"{supply_rmse[i]:.2f}%", ha="center", fontsize=10)
    plt.text(x[i] + bar_width/2, demand_rmse[i] + 0.02, f"{demand_rmse[i]:.2f}%", ha="center", fontsize=10)

# Display the plot
plt.show()
```
